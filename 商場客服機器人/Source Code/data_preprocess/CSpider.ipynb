```python
cspider_data = [ {'db_id': 'sports_competition',
  'question': '哪些位置既有得分超过20又有得分低于10的球员 ？',
  'sql': {'orderBy': [],
   'from': {'table_units': [['table_unit', 2]], 'conds': []},
   'union': None,
   'except': None,
   'groupBy': [],
   'limit': None,
   'intersect': {'orderBy': [],
    'from': {'table_units': [['table_unit', 2]], 'conds': []},
    'union': None,
    'except': None,
    'groupBy': [],
    'limit': None,
    'intersect': None,
    'where': [[False, 4, [0, [0, 18, False], None], 10.0, None]],
    'having': [],
    'select': [False, [[0, [0, [0, 13, False], None]]]]},
   'where': [[False, 3, [0, [0, 18, False], None], 20.0, None]],
   'having': [],
   'select': [False, [[0, [0, [0, 13, False], None]]]]},
  'query': 'SELECT POSITION FROM player WHERE Points  >  20 INTERSECT SELECT POSITION FROM player WHERE Points  <  10'}]

for cspider_dict in cspider_data:
    temp = {"phase":1,"table_id":"", "question":"","sql":""}
    for key,val in cspider_dict.items():
#         print(key,val)
        
        if(key == "db_id"):
            if 'UNION' in cspider_data[0]['query']:
                temp["table_id"] = cspider_data[0]["db_id"] + '['+ str(cspider_data[0]["sql"]['from']['table_units'][0][1]) + ']' + '[' + str(cspider_data[0]["sql"]['union']['from']['table_units'][0][1]) + ']'
            else:
                temp["table_id"] = cspider_data[0]["db_id"] + '['+ str(cspider_data[0]["sql"]['from']['table_units'][0][1]) + ']'            
        elif(key == "question"):
            temp["question"] = cspider_data[0]["question"]
        
        
        elif(key == "sql"):
            selAry = []
            if 'INTERSECT' in cspider_data[0]['query']:
                for col in val['intersect']["select"][1]:
                    selAry.append(col[1][1][1])
                for col in val["select"][1]:
                    selAry.append(col[1][1][1])
            elif 'UNION' in cspider_data[0]['query']:
                for col in val['union']["select"][1]:
                    selAry.append(col[1][1][1])
                for col in val["select"][1]:
                    selAry.append(col[1][1][1])
            else:
                for col in val["select"][1]:
                    selAry.append(col[1][1][1])
                    
                    
            condsAry = []           
            if 'INTERSECT' in cspider_data[0]['query']:
                for conds in val['intersect']["where"]:
                    condsAry.append([
                        conds[2][1][1],
                        conds[1],
                        conds[3]
                     ])
                for conds in val["where"]:
                    condsAry.append([
                        conds[2][1][1],
                        conds[1],
                        conds[3]
                     ])
            else:
                for conds in val["where"]:
                    if len(conds) > 4: # 排除or / and 
                        condsAry.append([
                            conds[2][1][1],
                            conds[1],
                            conds[3]
                             ])
                    else:
                        pass
                
            aggAry = []
            for col in val["select"][1]:
                aggAry.append(col[0])
                
            temp["sql"]= {
                  "sel":selAry,
                  "conds":condsAry,
                  "agg":aggAry
               }
            for j in range(len(temp['sql']['conds'])):
                if temp['sql']['conds'][j][1] == 2:
                    temp['sql']['conds'][j][1] = 0
                elif temp['sql']['conds'][j][1] == 3:
                    temp['sql']['conds'][j][1] = 1
                elif temp['sql']['conds'][j][1] == 4:
                    temp['sql']['conds'][j][1] = 2
                elif temp['sql']['conds'][j][1] == 7:
                    temp['sql']['conds'][j][1] = 3
print(temp)
```
