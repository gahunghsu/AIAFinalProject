```python
# op_sql_dict = {0:">", 1:"<", 2:"==", 3:"!="}
# agg_sql_dict = {0:"", 1:"AVG", 2:"MAX", 3:"MIN", 4:"COUNT", 5:"SUM"}
```

```python
# wikiSQL
# agg_ops = [0:'', 1:'MAX', 2:'MIN', 3:'COUNT', 4:'SUM', 5:'AVG']
# cond_ops = [0:'=', 1:'>', 2:'<', 3:'OP']
```

```python
# agg:
#     1 -> 5
#     2 -> 1
#     3 -> 2
#     4 -> 3
#     5 -> 4
```

```python
# cond:
#     0 -> 1
#     1 -> 2
#     2 -> 0
```

```python
# 在每个字典前面加入'phase': 1
data = [{'phase': 1, **d} for d in data]
```

```python
for item in data:
    # Modify agg
    agg_mapping = {1: 5, 2: 1, 3: 2, 4: 3, 5: 4}
    item['sql']['agg'] = [agg_mapping[x] if x in agg_mapping else x for x in item['sql']['agg']]
    # Modify conds
    cond_mapping = {0: 1, 1: 2, 2: 0}
    for cond in item['sql']['conds']:
        cond[1] = cond_mapping[cond[1]] if cond[1] in cond_mapping else cond[1]
```

```python
with open("final_train_data.json", "w") as train_data_file:
    json.dump(data, train_data_file, indent=4)
```

```python
import json
import sqlite3

# 讀取JSON文件
file = open('train.tables.json','r', encoding='utf-8')
data = []
for line in file.readlines():
    dic = json.loads(line)
    data.append(dic)
    
# 提取表名列表
table_names = data['id']

# 連接到資料庫
conn = sqlite3.connect('train.db')
cursor = conn.cursor()

# 遍歷表名列表，執行SQL查詢
for table_name in table_names:
    query = f"SELECT * FROM `{table_name}`"
    cursor.execute(query)
    # 檢索結果
    results = cursor.fetchall()
    # 逐行打印結果
    for row in results:
        print(row)

# 關閉連接
conn.close()
```
